## Reserved Words

Reserved words are identifiers that have a special meaning in a programming language. They are predefined and intended for particular use cases or functions. Because of their specific role, they cannot be used as identifiers (such as variable names, function names, or class names).
- Reserved words have specific meaning, use, or function defined by the language.
- Examples include: True, False, None, if, else, class, def, return, while, for, etc.
- These words are not allowed to be redefined or used as custom identifiers such (variable, function names)


## Keywords vs Reserved Words in Python

### 🔑 Keywords
- Keywords are **predefined words** in Python that define the **syntax, structure, and rules** of the language.  
- They are part of the grammar and control flow of Python programs.  
- Examples: `if`, `else`, `while`, `for`, `class`, `def`, `return`, `import`, `try`, `except`.  
- **Purpose:** They tell Python *how* to interpret and execute code.  

```python
# Example of keywords in use
if True:
    print("Hello, Python!")
```

---

### 📌 Reserved Words
- Reserved words are words that are **reserved for a particular meaning or use** in Python.  
- They cannot be used as identifiers (variable names, function names, class names) because they would **conflict** with Python’s internal usage.  
- Examples: `True`, `False`, `None`, `and`, `or`, `not`.  
- **Purpose:** They prevent naming conflicts and ensure consistency in the language.  

```python
# Invalid usage (reserved word as variable name)
None = 5   # ❌ SyntaxError
```

---

✅ **Summary:**  
- **Keywords** → Define the **rules and structure** of Python code.  
- **Reserved Words** → Reserved for **specific meanings** to avoid conflicts.


## Multiline Comments in Python

Python does not have a dedicated syntax for multiline comments like some other languages (e.g., `/* ... */` in C/Java). Instead, you can achieve multiline comments in two common ways:

### 1. Using Multiple `#` Symbols  
You can place the `#` symbol at the beginning of each line to create a block of comments.

```python
# This is line 1 of a comment
# This is line 2 of a comment
# This is line 3 of a comment
```

---

### 2. Using Triple Quotes (`''' ... '''` or `""" ... """`)  
Triple-quoted strings can also be used as multiline comments. Although technically they are **string literals**, if not assigned to a variable, Python ignores them, so they act like comments.

```python
'''
This is a multiline comment
written using triple single quotes
'''
```

```python
"""
This is another multiline comment
written using triple double quotes
"""
```

---

✅ **Summary:**  
- Use `#` for single-line or multiple consecutive lines.  
- Use `''' ... '''` or `""" ... """` for quick multiline comments or documentation-like notes.


## Rules for Defining Variables

1. A variable name must **start with a letter** (A–Z or a–z) or an **underscore (_)**.  
2. A variable name can contain **letters, digits, and underscores**.  
3. A variable name **cannot contain spaces**.  
4. Variable names are **case-sensitive** (e.g., `Name` and `name` are different).  
5. Variable names **cannot include special characters** such as `@ # $ % ^ & * ( )` — only the underscore `_` is allowed.  
6. **Reserved keywords** (e.g., `if`, `def`, `class`, etc.) cannot be used as variable names.  
7. An underscore can appear at the **beginning, middle, or end** of a variable name.  
8. Do not use the names of **built-in functions** (e.g., `print`, `help`, `sum`, `max`, `min`, `len`, `var`, etc.) as variable names.


## 🚫 Do Not Use Built‑in Functions as Variable Names

Python provides many built‑in functions. Avoid using these names for your own variables, functions, or classes.

### Common Built‑in Functions
| Function | Purpose |
|----------|---------|
| `print()` | Displays output |
| `input()` | Reads user input |
| `len()` | Returns length of an object |
| `sum()` | Returns sum of iterable |
| `max()` | Returns maximum value |
| `min()` | Returns minimum value |
| `sorted()` | Returns sorted list |
| `range()` | Generates a sequence of numbers |
| `type()` | Returns type of object |
| `id()` | Returns unique identifier of object |
| `help()` | Displays help documentation |
| `dir()` | Lists attributes of object |
| `abs()` | Absolute value |
| `round()` | Rounds number |
| `pow()` | Exponentiation |
| `divmod()` | Returns quotient & remainder |
| `all()` | True if all elements are true |
| `any()` | True if any element is true |
| `map()` | Applies function to iterable |
| `filter()` | Filters iterable |
| `zip()` | Combines iterables |
| `open()` | Opens a file |
| `eval()` | Evaluates string as Python code |
| `exec()` | Executes Python code |
| `globals()` | Returns global namespace |
| `locals()` | Returns local namespace |

---

### ❌ Bad Practice
```python
sum = 100      # Overrides built-in sum()
print = "Hi"   # Overrides built-in print()
len = 50       # Overrides built-in len()
```

### ✅ Good Practice
```python
total_sum = 100
message = "Hi"
length_value = 50
```

---

### 🔑 Key Rule
- Always choose **descriptive names** (e.g., `student_count`, `average_score`) instead of built‑in names.  
- If you accidentally override a built‑in, you can restore it by restarting your Python session or using `del` to remove your variable.




# 🔄 Python vs C Language – Key Differences

| Aspect | Python | C |
|--------|--------|---|
| **Level** | High-level programming language | Middle-level programming language |
| **Typing** | Dynamically typed (no need to declare variable types) | Statically typed (variables must be declared with type) |
| **Compilation/Execution** | Interpreted (runs line by line) | Compiled (translated into machine code before execution) |
| **Syntax** | Simple, readable, close to English | Complex, requires more boilerplate code |
| **Speed** | Slower (due to interpretation and dynamic typing) | Faster (compiled, closer to hardware) |
| **Memory Management** | Automatic (Garbage Collection) | Manual (programmer manages memory using `malloc`, `free`) |
| **Use Cases** | Web development, data science, AI/ML, scripting | System programming, embedded systems, OS development |
| **Error Handling** | Runtime errors detected during execution | Compile-time errors detected before execution |
| **Portability** | Highly portable across platforms | Portable but requires recompilation for each platform |
| **Learning Curve** | Easier for beginners | Steeper, requires understanding of low-level concepts |
| **Standard Library** | Very extensive (built-in + third-party modules) | Smaller, more focused on system-level operations |

---

## ✅ Summary
- **Python** → Easy to learn, flexible, widely used for applications like AI, data science, and web development.  
- **C** → Powerful, fast, and closer to hardware, widely used for system-level programming and performance-critical applications.
