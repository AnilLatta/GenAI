
## 🖥️ How to Execute Code in Jupyter/Colab

### 1. **Shift + Enter**
- Executes the **current cell**.  
- Moves the cursor to the **next cell**.  
- If there is no next cell, it will **create a new empty cell** below.  
```python
print("Hello, World!")
```
👉 Press `Shift + Enter` → runs the code and jumps to the next cell.

---

### 2. **Ctrl + Enter**
- Executes the **current cell**.  
- Keeps the cursor in the **same cell** (does not move or create a new one).  
```python
x = 5
print(x * 2)
```
👉 Press `Ctrl + Enter` → runs the code but stays in the same cell.

---

### 3. (Bonus) **Alt + Enter**
- Executes the **current cell**.  
- Always **creates a new empty cell below**, regardless of whether one exists.  

---

## 🎨 Summary
- **Shift + Enter** → run + go to next cell (or create new).  
- **Ctrl + Enter** → run + stay in same cell.  
- **Alt + Enter** → run + always create new cell below.


## 🐍 Headings in Python

Unlike languages like HTML or Markdown, **Python does not have a built‑in concept of headings**. But you can create headings in your code or notebooks using:

---

### 1. 📋 Comments as Headings
Use `#` for single‑line headings:
```python
# ==========================
#   DATA LOADING SECTION
# ==========================
```

This is the most common way to visually separate sections in scripts.

---

### 2. 📘 Docstrings as Headings
Use triple quotes (`""" ... """`) for multi‑line headings or documentation:
```python
"""
==========================
   DATA PROCESSING MODULE
==========================
This module handles cleaning and transformation of raw data.
"""
```

Often used at the top of files, classes, or functions.

---

### 3. 📓 Markdown Headings (in Jupyter/Colab)
In notebooks, you can use **Markdown cells** to create headings:
```markdown
# Heading 1
## Heading 2
### Heading 3
```

This is how you make proper headings in Jupyter/Colab for readability.

---

### 4. 🎨 Visual Separation
You can also use repeated characters for emphasis:
```python
# ----------- CONFIGURATION -----------
# ----------- FUNCTIONS ---------------
# ----------- MAIN PROGRAM ------------
```
